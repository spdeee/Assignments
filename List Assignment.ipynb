### 1. Write a code to compute the sum of the two lowest numbers and the two highest numbers in the following list.

```python
import numpy as np
```

```python
my_list = [34, 56, 76, 45, 2, 12, 67, 98, 37, 54, 66]
list_sorted = np.sort(my_list)
print(list_sorted)
```

```python
# Sum of the two lowest numbers
min_sum = list_sorted[0] + list_sorted[1]
print('Sum of the two lowest numbers is:', min_sum)
```

```python
# Sum of the two highest numbers
max_sum = list_sorted[-1] + list_sorted[-2]
print('Sum of the two highest numbers is:', max_sum)
```

### 2. The following two lists contain student names and scores. Write a code that gets the name from the user and prints the score of that student. 


```python
names = ["David", "Michael", "John", "James", "Greg", "Mark", "William", "Richard", "Thomas", "Steven","Mary", "Susan", "Maria", "Karen", "Lisa", "Linda", "Donna", "Patricia", "Debra", "Eric"]
scores = [99, 87, 78, 86, 68, 94, 76, 97, 56, 98, 76, 87, 79, 90, 73, 93, 82, 69, 97, 98]
```

```python
print('Enter the name of a student')
name = input()
```

```python
# Get index of the student from names list and find corresponding score using the index
ind = names.index(name)
score = scores[ind]
```

```python
# Using the index to get the score 
print('The Score of the student {} is {}'.format(name, score))
```

### 3. By using the two lists above, what is the maximum score and how many students got that score?

```python
max_score = max(scores)
print('Maximum value of the score is:', max_score)
count_score = scores.count(max_score)
print('{} students got the maximum score'.format(count_score))
```
