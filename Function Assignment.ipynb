### 1. Finding Prime number

```python
def prime_number():
    
    """A function that ask for an integer and 
    determines if that integer is a prime number of not"""
        
    i = int(input('Enter an integer '))
    if i//2 == 0:
        print('{} is not a prime number'.format(i))
    elif i >1 :
        print('{} is a prime number'.format(i))
```

```python
prime_number()
```

### 4. Blue and red marbles

```python
def guess_blue(tb,tr,b,r):
   
    """A function that takes total number of blue marbles and total number of red 
    marbles as the first two inputs. It also takes number of blue and red marbles pulled 
    out so far as the next two inputs.Based on these 4 inputs the function gives the 
    probability of pulling a blue marble."""
    
    # Total number of marbles left in the bag
    tot_mar = (tb + tr) - (b + r)
    
    # Blue and red marbles still in bag
    b_bag = tb - b
    r_bag = tr - r
    
    # probability of picking a blue marble
    prob = b_bag/tot_mar
    
    return prob
```

```python
guess_blue(5,5,2,3)
```

### 5. Fillable

```python
def fillable(stock, item, n):

    """This function takes the merchandise in stock as a dictionary, 
    a particular item that a customer wants to buy as a string and 
    number of those items (>=1) the customer wants to buy, as arguments. It 
    returns True if those items are fillable else returns False"""
    
    #items in stock
    in_stock = stock[item]
    
    return in_stock >= n
```

```python
stock = {'mugs':25, 'tshirts':40, 'pens':50}
```

```python
fillable(stock, 'tshirts', 45)
```

```python
fillable(stock, 'mugs', 10)
```

### 6. Write a function called take_umbrella that takes two arguments: a string representing the current weather, and a float representing the chance of rain today.

```python
def take_umbrella(weather, rain):
    
    """This function takes the current weather(string) and chance of
    rain(float) as arguments and outputs if one should carry an umbrella """
    
    # We need an umbreall when the weather is rainy or if the chance of rain is 
    # more than 70%
    
    if weather == 'Rainy' or rain >=0.7 :
        print('Take an umbrella')
    else:
        print('Do not take an umbrella')
```

```python
take_umbrella('Rainy', 0.8)
```

```python
take_umbrella('Cloudy', 0.65)
```

```python
take_umbrella('Sunny', 0.8)
```

```python
take_umbrella('Cold', 0.5)
```
