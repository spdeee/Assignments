### 1. Create three lists representing house features, each list containing ten values. The first one for the house's size in square meters, the second one for rooms and last for price. Then, create an array combining these lists.


```python
house_size = [100, 300, 250, 618, 150, 412, 190, 290, 115, 275]
house_rooms = [2, 4, 4, 7, 3, 5, 3, 5, 2, 4]
house_price = [90000, 200000, 175000, 700000, 125000, 350000, 150000, 195000, 100000, 185000]

hou_mar = np.array([house_size, house_rooms, house_price])
hou_mar
```

### 2. Transpose the array you have created, so that every line can represent features of one house.

```python
houses = np.transpose(hou_mar)
houses
```

### 3. Display the shape of the array and explain what it means.


```python
houses.shape
```

#### The initial array was a 3 x10 matrix, each 'row' was representing one feature of the houses. By transposing we turned it into a 10 X 3 matrix where each row now represents all the features of a single house.
