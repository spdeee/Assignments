### 2. Notice that the data has a time dimension (year). For this assignment, forget about time and treat all the observations as if they're from the same year. Choose a strategy to deal with the missing values for each variable. For which variables would filling in the missing values with some value make sense? From the above result, it can be seen that there are missing values in almost all the variables of the dataframe. This fraction is too high to ignore in most of the variables, except for the variables where it is less than 10%. For these variables we will drop the missing values and for the rest of the variables we will fill the missing values with the mean value of the entire column (ignoring the year).

```python
# Columns where the missing value fraction is less than 10% and are dropped
cols_drop_vals = ['GRADES_KG_G','GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G','GRADES_ALL_G']

# Columns to fill in the missing values with mean (ignoring the year)
cols_fill = ['ENROLL', 'TOTAL_REVENUE','FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE','TOTAL_EXPENDITURE',
             'INSTRUCTION_EXPENDITURE','SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE', 
             'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G','GRADES_1_8_G','GRADES_9_12_G', 'AVG_MATH_4_SCORE', 
             'AVG_MATH_8_SCORE','AVG_READING_4_SCORE', 'AVG_READING_8_SCORE' ]
```

```python
# Making a copy of the dataframe 
states_df_copy1 = states_df.copy()
```

```python
#Dropping columns with missing value fraction less than 10
states_df_copy1 = states_df_copy1.dropna(subset = cols_drop_vals)
```

```python
# The missing value fraction after dropping them from the selected columns
(states_df_copy1.isnull().sum()*100/states_df_copy1.shape[0]).sort_values(ascending=False)
```

```python
#Dropping the missing values from selected columns does not affect the data by a lot
states_df_copy1.describe()
```

```python
# A function to fill missing values with means of the variables (ignoring the years)

def fill_cols(df, cols): 
    
    """This function takes a dataframe and a set of columns. It 
    fills the missing values of columns 
    in cols list with their mean value"""
    
    for i in cols :
        df[i].fillna(df[i].mean(), inplace = True)

fill_cols(states_df_copy1, cols_fill)
```

```python
states_df_copy1.describe()
```

```python
# All the missing values from all the columns have been filled
(states_df_copy1.isnull().sum()*100/states_df_copy1.shape[0]).sort_values(ascending=False)
```

### 3. Now, take into account the time factor. Replicate your second answer, but this time, fill in the missing values by using a statistic that is calculated within the year of the observation. For example, if you want to fill in a missing value for a variable with the mean of that variable, calculate the mean by using only the observations for that specific year.

```python
# Making another copy of the dataframe to work with
states_df_copy2 = states_df.copy()
```

```python
(states_df_copy2.isnull().sum()*100/states_df_copy2.shape[0]).sort_values(ascending=False)
```

```python
# Dropping the missing values from columns where the fraction is less than 10 
states_df_copy2 = states_df_copy2.dropna(subset = cols_drop_vals)
(states_df_copy2.isnull().sum()*100/states_df_copy2.shape[0]).sort_values(ascending=False)
```

```python
# Calculating the mean of each variable per year
means = states_df_copy2.groupby('YEAR').mean()
years = list(means.index)
```

```python
# Filling the missing value of each vairiable with the mean for that specific year

for y in years:
    for i in range(len(states_df_copy2)):
        if states_df_copy2['YEAR'][i] == y:
            for j in cols_fill:
                states_df_copy2[j].fillna(means.loc[y,j], inplace = True)
        
```

```python
states_df_copy2.describe()
```

```python
# Checking to make sure all the missing values have been filled
(states_df_copy2.isnull().sum()*100/states_df_copy2.shape[0]).sort_values(ascending=False)
```

### 5. This time, fill in the missing values using interpolation (extrapolation)

```python
states_df_copy3 = states_df.copy()
```

```python
(states_df_copy3.isnull().sum()*100/states_df_copy3.shape[0]).sort_values(ascending=False)
```

```python
# Dropping the missing values from columns where the fraction is less than 10 
states_df_copy3 = states_df_copy3.dropna(subset = cols_drop_vals)
(states_df_copy3.isnull().sum()*100/states_df_copy3.shape[0]).sort_values(ascending=False)
```

```python
# Filling the missing values with the interpolation for selected variable
for i in cols_fill:
    states_df_copy3[i].interpolate(method ='linear', limit_direction ='forward', inplace=True)
```

```python
states_df_copy3.describe()
```

```python
# Missing value fraction after applying interpolation
(states_df_copy3.isnull().sum()*100/states_df_copy3.shape[0]).sort_values(ascending=False)
```

It can be noted from the above table that the interpolation method does not eliminate the missing values completely for some variables. This makes sense as this method uses previous values to 'calculate' the missing value based on where the missing values are present in the data, this method may not have enough values to calculate the missing value.
