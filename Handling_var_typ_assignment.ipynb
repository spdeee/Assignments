### 2. Using the movie data, we have studied in this lesson. Complete the following tasks:

#### Select a variable from the movie dataset and discuss the variable type of this variable

The variable 'Genre' is of type categorical, as it is used to let us know which category a movie belongs to. It is a nominal categorical variable, as we cannot sort or rank the genre of a movie. 

#### Categorize a variable of your choice and assign a name to it

```python
def length_movie(runtime):
   
    """ This function takes the length of a movie in minutes 
    and categorizes it into three groups, Long, Normal and Short"""
    
    if runtime <= 100:
        return 'Short'
    
    if (runtime > 100) and (runtime<125):
        return 'Normal'
    
    if runtime >=125 :
        return 'Long'
    
    
movie_df['Length'] = movie_df['Runtime (Minutes)'].apply(length_movie)

movie_df.head()
```

#### Count the number of rows for each category

```python
movie_df.groupby('Length')['Length'].count()
```

### 3.  Is it possible to convert categorical variable into numerical one? If so, how can you make that?

It is possible to convert categorical variables into numerical one by assigning a number to each category. The numerical value of the assigned number is based on the type of the categorical variable. A converted nominal categorical variable has no numerical meaning as it does not represent any quality/attribute of the data. A converted ordinal categorical data on the other hand can be sorted or represent some gradation.  

### 4. While continuous data is generally preferred over discrete data, please indicate circumstances where discrete is the preferred data type although continuous data is available for the same characteristic.

One example of continuous data that is preferable in discrete form is age of a person. Age is a continuous data type, as our age changes every day, every minute matter of fact. So even if the age is present in continuous form, a discrete value is preferred. For example, it is preferred to state age as 10 years as opposed to 5.25e+6 minutes.
