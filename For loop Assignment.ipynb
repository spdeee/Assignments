### 1. Do the following to create a program that simulates how websites ensure that everyone has a unique username.
* Make a list of five or more usernames called current_users.
* Make another list of five usernames called new_users. Make sure one or two of the new usernames are also in the current_users list.
* Loop through the new_users list to see if each new username has already been used. If it has, print a message that the person will need to enter a new username. If a username has not been used, print a message saying that the username is available.
* Make sure your comparison is case insensitive. If 'John' has been used, 'JOHN' should not be accepted.

```python
current_users = ['John', 'Matthew', 'Rosy', 'AvA', 'Mary']
new_users = ['Mona', 'Barry', 'MATTHEW', 'Ava', 'Greg']

#Converting the current usernames to lower case
current_users_lower = []
for i in range(len(current_users)):
    current_users_lower.append(current_users[i].lower())
print(current_users_lower)
```

```python
for i in range(len(new_users)):
    if new_users[i].lower() in current_users_lower:
        print('{} user name is not available, enter a new name'.format(new_users[i]))
    else:
        print('{} user name is available'.format(new_users[i]))
```

### 4. process_data Function

```python
def process_data(data):
    
    """ A function that takes a list with sublists, 
    subtracts the elements of each sublist and gives 
    the product of the results of the subtraction"""
    
    #Subtracting the sublist items
    sub = []
    for i in range(len(data)):
        ele = data[i][0] - data[i][1]
        sub.append(ele)

    #Multiplying the elements of the differences
    x = 1
    for i in sub:
        x = x*i
    return x
```

```python
data = [[2, 5], [3, 4], [8, 7]]
process_data(data)
```

### 5. user_contacts Function

```python
def user_contacts(data):
    
    """A function that takes a list of usernames and zipcodes; 
    converts it to a dictionary with username as key and 
    zipcode as value"""
    
    final_dict = {}
    for i in range(len(data)):
        if len(data[i]) == 2 :
            final_dict[data[i][0]] = data[i][1]
        else :
            data[i].append('None')
            final_dict[data[i][0]] = data[i][1]
    return final_dict
```

```python
data = [["Grae Drake", 98110], ["Bethany Kok"], ["Alex Nussbacher", 94101], ["Darrell Silver", 11201]]
user_contacts(data)
```
