### 11. Merge these two DataFrames into a single DataFrame called census. Please use the following dataset for merging purposes:

```python
df1 = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')

df2 = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')
```

```python
census = df1.merge(df2)
census.head()
```

### 12. Re-shape census so that one column contains all population measures, and another the year attributes.

```python
cen = census.melt(id_vars=['state', 'region', 'division'], var_name='year')
cen.head()
```

### 13. Group the data by year and summarize it.

### 14. Group the data by region, division and year and summarize it.
