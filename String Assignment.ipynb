### 2. Ask the user to type in the first line of a poem Raven by Edgar Allen Poe and display the length of the string. Ask for a starting number and an ending number and then display just that section of the text (remember Python starts counting from 0 and not 1).

```python
print('Enter the first line of the poem Raven by Edgar Allen Poe')
line=input()
```

```python
print ('Length of the first line:', len(line))
```

```python
print('Enter a starting number')
x = input()
x= int(x)-1
```

```python
print('Enter the ending number')
y = input()
y=int(y)-1
```

```python
print('Text between starting and ending number is:', line[x:y])
```
