```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
studperf_df = pd.read_csv('StudentsPerformance.csv')
```

```python
studperf_df.head()
```

### 1.Does the performance in the exams vary according to gender, origin and education level of parents?

#### Let's check if the performance is affected by these variables by checking the mean of performace for each variable

```python
studperf_df.groupby('gender').mean()
```

```python
studperf_df.groupby('race/ethnicity').mean()
```

```python
studperf_df.groupby('parental level of education').mean()
```

From the above observations, it can be seen that performance does vary based on gender and origin and level of parents education.

### 2. Does lunch type have a relationship with exam performances? How can you explain this, if any?

```python
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
sns.barplot(studperf_df["lunch"], studperf_df["math score"])
plt.title("Math Score")
plt.subplot(1,3,2)
sns.barplot(studperf_df["lunch"], studperf_df["reading score"])
plt.title("Reading Score")
plt.subplot(1,3,3)
sns.barplot(studperf_df["lunch"], studperf_df['writing score'])
plt.title("Writing Score")
plt.show()
```

#### Students with 'standard' lunch type fair better than with those with 'free/reduced' type. One plausible explanation is that the students with 'free/reduced' lunch option come from weak economical background and have no means to prepare for exams

### 3. Does the preparation courses have an impact on exam performance?

```python
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
sns.barplot(studperf_df["test preparation course"], studperf_df["math score"])
plt.title("Math Score")
plt.subplot(1,3,2)
sns.barplot(studperf_df["test preparation course"], studperf_df["reading score"])
plt.title("Reading Score")
plt.subplot(1,3,3)
sns.barplot(studperf_df["test preparation course"], studperf_df['writing score'])
plt.title('Writing Score')
plt.show()
```

#### Yes, students that took preparation courses have done better on all the exams compared to the ones who did not take these courses

### 4. Which lessons are most correlated with each other?

```python
plt.figure(figsize=(18,10))
corr_stu=studperf_df.corr()
sns.heatmap(corr_stu, annot=True, fmt='.2f', annot_kws={"size": 11}, linewidths=.5, vmin=0, vmax=1, cmap='viridis')
plt.title("Correlation ")
plt.show()
```

```python
corr_stu
```

#### From the above graphs it can be seen that reading and writing are most correlated, which makes sense - if you are better at reading, it is ver well possible that you will be better at writing too
